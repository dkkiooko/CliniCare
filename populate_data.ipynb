```python
# get permutations of dates in blocks and combine them
def get_perms_10000(arg, total=10000):
    # get permutations and combine them
    args_list = []
    for i in range(1500//365 + 1):
        args_list = np.concatenate((args_list, np.random.permutation(arg)))
    args_list = args_list[:1500]
    return args_list
```
