```python
for i in range(doctors.shape[0]):
    print("INSERT INTO doctors (id, created_at, updated_at, doctor_id, fname, lname, office_number, specialty) values (", end="")
    for key, value in doctors.loc[i,:].items():
        if type(value) == str and i != doctors.shape[0] -1:
            print('"'+value+'" ', end=', ')
        elif type(value) == str and i == doctors.shape[0] -1:
            print('"'+value+'"', end="")
        elif type(value) != str and i == doctors.shape[0] -1:
            print(value, end='')
        else:
            print(value, end=', ')
    print(");")
    print('\n\n\n')
```
